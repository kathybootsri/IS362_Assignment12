### Step 1: Created API and Used URL for Hardcover-Fiction List

```python
import pandas as pd
import requests
import json

URL = "https://api.nytimes.com/svc/books/v3/lists/current/hardcover-fiction.json?api-key=a7AawEETZWlcHTDiYlsAYFnqrS3SXlaK"
```

### Step 2: Retrieve Page with Requests and Pretty Print Page

```python
page = requests.get(URL) 

json_page = page.json()

print(json.dumps(json_page, indent=1))
```

### Step 3: Create Dataframe from Results Dictionary, Books key of dictionary

```python
results = page.json()['results']

books = results['books']

df = pd.DataFrame(books)

print(df.head())
```

### Step 4: Append List Details to Dataframe

```python
for row in range(len(df)):
    df['List Name'] = results['list_name']
    df['Bestsellers Date'] = results['bestsellers_date']
    df['Published Date'] = results['published_date']
    df['List Frequency'] = results['updated']
    
print(df)
```

### Step 5: Dump Column to Create Print Statement to Retrieve Bestselling Book

```python
for x in range(len(df.columns.values)):
    print(f"{x}: {df.columns.values[x]}")
```
